# Iris Dataset and Analysis

```python
#import seaborn 
import seaborn as sns 
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris.describe
```

```python
iris.info
```

```python
sns.pairplot(data=iris,hue='species')
```

```python
#Boxplot 
sns.boxplot(x='species', y='sepal_length',data=iris,hue='species')
```

```python
sns.boxplot(x='species',y='sepal_width', data=iris,hue='species')
```

```python
sns.boxplot(x='species',y='petal_length', data=iris,hue='species')
```

```python
sns.boxplot(x='species', y='petal_width',data=iris,hue='species')
```

```python
#Violinplot
sns.violinplot(x='species',y='petal_length',data=iris,hue='species')
```

```python
iris.hist(edgecolor='black', linewidth=1.2, figsize=(10, 8))
```

```python
#correlation 
numeric_corr = iris.select_dtypes(include=['number'])
```

```python
corr = numeric_corr.corr()
```

```python
sns.heatmap(corr, annot=True, cmap='coolwarm')
```

```python
sns.scatterplot(x='sepal_width',y='sepal_length', data=iris,hue='species')
```

```python
sns.scatterplot(x='petal_width',y='petal_length',data=iris,hue='species')
```
